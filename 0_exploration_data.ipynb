```python
data = read_json("data_videos\data_videos.json")
df = pd.DataFrame(data)
df.head()
```

```python
df.info()
```

```python
df = df[['titulo_video','nombre_canal' ]]
```

```python
df_per_canal = df.nombre_canal.value_counts()
df_per_canal = df_per_canal.to_frame().reset_index() 
df_per_canal.head(10)
```

```python
df_per_canal[:20].plot.scatter(x='nombre_canal',
                                 y='count',
                                 rot="vertical",
                                 grid= True,
                                #  logy = True, 
                                #  x_compat=True
                                 )
```

```python
n = 10
df_per_canal[:n].plot.pie(labels=df_per_canal[:n].nombre_canal, 
                           y='count',
                           fontsize=10,
                           autopct="%.2f",
                           legend = False, 
                           figsize=(6, 6))
```
